# Removing outlier

```python
from scipy.stats import zscore
z=zscore(df)
z
```

```python
df_new=df[(z<3).all(axis=1)]
```

```python
x= df.drop(['W'],axis=1)
y=df['W']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.22,random_state=99)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

# MODELS

```python
#Linear Regression

from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
print(LR.score(x_train,y_train))
LR_predict=LR.predict(x_test)
```

```python
print(mean_squared_error(LR_predict,y_test))
print(mean_absolute_error(LR_predict,y_test))
print(r2_score(LR_predict,y_test))
```

```python
#SVR

from sklearn.svm import SVR
svr=SVR(kernel='linear')
svr.fit(x_train,y_train)
print(svr.score(x_train,y_train))
svr_predict=svr.predict(x_test)
```

```python
print(mean_squared_error(svr_predict,y_test))
print(mean_absolute_error(svr_predict,y_test))
print(r2_score(svr_predict,y_test))
```

```python
#Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
print(rf.score(x_train,y_train))
rf_PRED=rf.predict(x_test)
```
